####  __iv.__ Find the matrix representation $Amtx$ of the linear map A. What is the size of the matrix $Amtx$? Plot the 0-th, 15-th, 31-th and 63-th columns. Do you see any relation between them? Note: Here we used Python’s numbering convention, starting from 0.


```python
fig, axs = plt.subplots(4,figsize=(8,15))

aux=[]
count=0
for i in x:
    z=np.zeros((64,1))
    z[i]=1
    aux.append(A(z))
    if i == 0 or i==15 or i==31 or i==63:
        axs[count].plot(x,A(z),c='g')
        axs[count].set_title(str(i)+'th column')
        count+=1
fig.suptitle('Amtx columns')
Amtx=np.round(np.array(aux).T,2)
```

```python
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', None)  
pd.DataFrame(Amtx)
```

####  __v.__ Find the matrix representation $Amtx$ of the linear map A. What is the size of the matrix $Amtx$? Plot the 0-th, 15-th, 31-th and 63-th columns. Do you see any relation between them? Note: Here we used Python’s numbering convention, starting from 0.


```python
ones=np.ones((64,1))
ones[0:-1:2]=-1
x=np.arange(0,64)
plt.plot(x,ones,label='z')
plt.plot(x,A(ones),c='m',label='A(z)')
plt.title('Convolution')
plt.xlabel('ith component')
plt.ylabel('value ith component')
plt.legend()
plt.show()
```

####  __vi.__  Download the file hw03\_signal\_3.npz containing the vector $x_3$. Is this vector on the range of the matrix $Amtx$? Hint: Try to formulate this question in terms of a least-squares problem.

```python
f1=np.load('hw03_signal_3.npz')['arr_0']
f2=A(f1)
x=np.arange(0,64)
plt.plot(x,f1,c='m',label='x1')
plt.plot(x,f2,label='A(x1)')
plt.title('Convolution')
plt.xlabel('ith component')
plt.ylabel('value ith component')
plt.legend()
plt.show()
```

```python
print('residues:\n', np.linalg.lstsq(Amtx,f1)[-1])
```

#### __vii.___ Find the vector in the range of the matrix Amtx that is closest in Euclidean distance to x3. Plot this vector.

```python
xhat= np.linalg.lstsq(Amtx,f1)[0]
```

```python
print('x most closer is:\n', xhat)
```

```python
plt.plot(x,xhat,c='m',label='x3')
plt.title('Convolution')
plt.xlabel('ith component')
plt.ylabel('value ith component')
plt.legend()
plt.show()
```
